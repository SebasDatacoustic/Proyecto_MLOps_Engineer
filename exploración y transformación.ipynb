```python
# Vamos a crear una función para el retorno de inversión
def retorno_inversion(revenue, budget):
    if revenue is None or budget is None or budget == 0:
        return 0
    else:
        return revenue / budget
```

### Función III. Esta función me devuelve el nombre de la película, el año de estreno y el puntaje de popularidad


```python
#antes de crear la función he decidido capitalizar la columna de los títulos por medio de la siguiente función.
# uso str.split para dividir cada nombre en palabras individuales, luego a esa lista le paso
#una funciçon lambda que usa la función capitalize py luego con un join se vuelven a unir con un espacio entre ellas.

movies_dataset['title'] = movies_dataset['title'].str.split().apply(lambda x: ' '.join(word.capitalize() for word in x))
```

```python
# en esta función se compara el título que ingresa el usuario y si coincide nos devuelve 
# el primer valor (values[0]) de la serie generada para título, año de estreno, y popularidad.
# En el caso de que NO HAYA RESULTADO para el título ingresado se le dice que no se encontró la película

def score_titulo(titulo_de_la_filmacion):
    titulo_de_la_filmacion = titulo_de_la_filmacion.title()
    
    
    
    pelicula = movies_dataset.loc[movies_dataset['title'] == titulo_de_la_filmacion]
    if pelicula.empty:
        return  "No se encontró la película."
    
    else:
        titulo = pelicula['title'].values[0]
        año_estreno = pelicula['release_year'].values[0]
        score = pelicula['popularity'].values[0]
        return f"La película {titulo} fue estrenada en el año {año_estreno} con un score de popularidad de {score}."
```

### Función IV. recibe el nombre de una película y devuelve la votación y el promedio

```python
# aquí se compara el título ingresado por el usuario, y si coincide entonces nos devuelve
# total de vots, año de estreno, promedio de votos y se le pone un condicional en caso de que el
# número de votos sea mayor o inferior a 2000
def votos_titulo(titulo_de_la_filmacion):
    titulo_de_la_filmacion = titulo_de_la_filmacion.title()
    pelicula = movies_dataset.loc[movies_dataset['title'] == titulo_de_la_filmacion]
    
    if pelicula.empty:
        return "No se encontró la película."
    
    total_votos = pelicula['vote_count'].values[0]
    promedio_votos = pelicula['vote_average'].values[0]
    año_estreno = pelicula['release_year'].values[0]
    
    if total_votos >= 2000:
        return f"La película {titulo_de_la_filmacion} fue estrenada en el año {año_estreno}. La misma cuenta con un total de {total_votos} valoraciones, con un promedio de {promedio_votos}."
    else:
        return "La película no cumple con el requisito de tener al menos 2000 valoraciones."
```

```python
votos_titulo('Batman')
```

### Función V. recibe el nombre de un actor  y devuelve la participación el número de películas en las que ha participado, además del retorno y el promedio por filmación.


```python
# para esta función voy a trabajar con los datasets que he guardado para unirlos
# usando el tipo de fusión outer, para qeu se incluyan todas las filas así no haya coincidencia exacta
# y ponga nan a los que no tengan coincidencia. 
movies_dataset_completo = pd.merge(movies_dataset_listo, credits_listo, on='id', how='outer')
```

```python
movies_dataset_completo.head(10)
```

```python
# ahora si podemos crear la función, y usamos str.contain qeu es una búsqueda de subcadena 
#con case= false nos aseguramos que no distinga entre mayúsculas y minúsculas
# además tuve que rellenar algunos valores faltantes con una cadena vacía
def get_actor(nombre_actor):
    movies_dataset_completo['Reparto'] = movies_dataset_completo['Reparto'].fillna('')   
    actor = movies_dataset_completo[movies_dataset_completo['Reparto'].str.contains(nombre_actor, case=False)]
    cantidad_filmaciones = actor.shape[0]
    éxito = actor['return'].sum()
    promedio = éxito / cantidad_filmaciones if cantidad_filmaciones > 0 else 0

    return f"El actor {nombre_actor} ha participado de {cantidad_filmaciones} filmaciones. El mismo ha conseguido un retorno de {éxito} con un promedio de {promedio} por filmación"
```

### Función VI. recibe el nombre de un Director  y devuelve el éxito a través del retorno, además el nombre de cada película,con fecha de lanzamientom retorno individual, costo y ganancia de esta.

```python
def get_director(nombre_director):
    movies_dataset_completo['Director'] = movies_dataset_completo['Director'].fillna('')
    
    director = movies_dataset_completo[movies_dataset_completo['Director'].str.contains(nombre_director, case=False)]
    if director.empty:
        return "No se encontraron películas del director."
    
    éxito = director['return'].sum()
    
    peliculas = director[['title', 'release_year', 'return', 'budget', 'revenue']].to_dict('records')

    return {
        'Éxito': éxito,
        'Películas': peliculas
    }
```

### Función VII. Esta es la función para recomendación de películas similares

```python
movies_dataset_recomendación= pd.read_csv("movies_dataset_recomendación.csv")
```

```python
# usamos la similitud del coseno, usando la función cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity


def recomendacion(titulo):
    
    
    
    pelicula_seleccionada = movies_dataset_recomendación[movies_dataset_recomendación['title'] == titulo]

    if pelicula_seleccionada.empty:
        return "No se encontró la película."

    #  matriz de simililitud entre películas para esto, usamos la fila de la peli sin el título y otras columnas qu eno necesito,
    #  además todo el dataset sin la columna tiítulo y las mismas columnas
    
    similitudes = cosine_similarity(pelicula_seleccionada.drop('title', axis=1), movies_dataset_recomendación.drop('title', axis=1))
    # accedemos con el argsort a las películas con similitud a la primera película que en este caso
    #  es la película que buscamos, y luego se seleccionan los útimos 5 índices (excluyendo la película
    # seleccionada) y por último se invierten los índices. 
    indices_similares = similitudes.argsort()[0][-6:-1][::-1]

    # finalmente obtenemos los títulos de las películas más similares y lo convertimos en lista
    peliculas_similares = movies_dataset_recomendación.iloc[indices_similares]['title'].tolist()

    return peliculas_similares
```
